### Removing Unnecessesary columns

```python
unnecessesary_columns = ['EmployeeCount', 'Education', 'Over18', 'StandardHours' ]
```

```python
df = df.drop(columns = unnecessesary_columns)
```

#### Eliminating redundant entries

```python
df.drop_duplicates(inplace=True)
```

#### Checking missing values

```python
df.dropna(inplace=True)
```

#### Exporting cleaned data 

```python
df.to_csv('cleaned_hr_data.csv', index=False)
```

#### Encode categorical variables 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Encoder = LabelEncoder()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
df[categorical_cols] = df[categorical_cols].apply(Encoder.fit_transform)
```
